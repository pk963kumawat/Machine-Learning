### Logistic Regression With ROC Curve 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.dummy import DummyClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
##sample binary classification data
X, y = make_classification(n_samples=1000, n_features=20, random_state=42)
```

```python
##Split into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
##Train Dummy and Logistic models
dummy_model = DummyClassifier(strategy='most_frequent')
dummy_model.fit(X_train, y_train)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
## Get predicted probabilities (positive class only)
dummy_model_prob = dummy_model.predict_proba(X_test)[:, 1]
model_prob = model.predict_proba(X_test)[:, 1]
```

```python
 ##Calculate FPR, TPR for ROC
dummy_fpr, dummy_tpr, _ = roc_curve(y_test, dummy_model_prob)
model_fpr, model_tpr, _ = roc_curve(y_test, model_prob)
```

```python
##Plot ROC Curve
plt.plot(dummy_fpr, dummy_tpr, linestyle='--', label='Dummy Model')
plt.plot(model_fpr, model_tpr, marker='.', label='Logistic')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```
