```python
import pandas as pd
import matplotlib.pyplot as plt
import  numpy as np
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
print(iris['DESCR'])
```

```python
print(iris['target'])
```

```python
## Independent Features
x=pd.DataFrame(iris['data'],columns=['sepal length in cm','sepal width in cm','petal length in cm','petal width in cm'])
```

```python
## Dependent Features
y=iris['target'] 
```

```python
## Train tast plot
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
## Apply Descision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
treeclassifier= DecisionTreeClassifier(max_depth=2)
```

```python
treeclassifier.fit(X_train,y_train)
```

```python
## Visualize the Decision Tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier, filled=True)
```

```python
y_pred=treeclassifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(classification_report(y_test, y_pred))
```

## Diabetes Predicition

```python
from sklearn.datasets import load_diabetes
```

```python
dataset = load_diabetes()
```

```python
print(dataset['DESCR'])
```

```python
dataset 
```

```python
import pandas as pd
df_diabetes = pd.DataFrame(dataset.data,columns=['age',                           
   'sex',
   'bmi',
   'bp',
   's1',
   's2',
   's3',
   's4',
   's5',
   's6'])
df_diabetes.head()
```

```python
## Independent  and Dependent Features
x = df_diabetes
y = dataset.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
X_train.head()
```

```python
## Correlation Matrix
X_train.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 8))
sns.heatmap(X_train.corr(),annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(X_train, y_train)
```

## Hpperparameter Tuning

```python
param={
    'criterion':['gini','entropy'],
    'splitter':['best','random'],
    'max_depth':[2,3,4,5,6,7,8,9,10],
    'max_features':[None,'auto','sqrt','log2'],
    }
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [3, 5, 10, None]
}

grid = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=5)
grid.fit(X_train, y_train)

print("Best Params:", grid.best_params_)
print("Best Score:", grid.best_score_)
```

```python
grid.best_params_
```

```python
## predictions
y_pred=grid.predict(X_test)
```

```python

from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```
